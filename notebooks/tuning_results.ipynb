```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
```

### Multi Horizon Output

```python
sho_gat = pd.read_parquet("../output_old/sho/study_results_gat/sho_gat_trials_df.parquet").sort_values("value")
sho_gcn = pd.read_parquet("../output_old/sho/study_results_gcn/sho_gcn_trials_df.parquet").sort_values("value")
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=sho_gat, x="params_graph_threshold", y="value")
plt.xlabel("Distance Threshold (km)")
plt.ylabel("Validation Score")
plt.title(f"GAT")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=sho_gcn, x="params_graph_threshold", y="value")
plt.xlabel("Distance Threshold (km)")
plt.ylabel("Validation Score")
plt.title(f"GCN")
plt.show()
```
